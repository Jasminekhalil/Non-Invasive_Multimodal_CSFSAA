```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import shap
import umap
import numpy as np
import matplotlib.patches as mpatches
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import HistGradientBoostingClassifier
from scipy.stats import spearmanr
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc, roc_curve
from imblearn.over_sampling import SMOTE
from sklearn.cluster import KMeans, SpectralCoclustering
```

```python
df = pd.read_csv("Gait_Data_with_Selected_Features.csv") # Reading the merged dataset we created

df = df[df["CSFSAA"].isin([0,1])].copy() # Creating a copy where CSFSAA status is valid
df["label"] = (df["CSFSAA"] == 1).astype(int) # Label column
y = df["label"]
id_cols = ["PATNO", "EVENT_ID", "INFODT", "CSFSAA"] 
df.drop(columns=id_cols, inplace=True) 

gait_features = [col for col in df.columns if any(keyword in col for keyword in ["SP_", "RA_", "LA_", "SYM_", "JERK_", "ASA", "TRA", "STR_", "STEP_", "TUG", "SW_", "ASYM", "CAD"])]
demographic_features = ["age", "age_at_visit", "agediag", "ageonset", "SEX", "EDUCYRS", "fampd", "fampd_bin"]
clinical_features = ["moca", "MCI_testscores", "cogstate", "gds", "ess", "stai", "stai_state", "stai_trait",
    "NP1ANXS", "NP1APAT", "NP1COG", "NP1DPRS", "NP1FATG", "NP1HALL", "NP1DDS",
    "HVLTFPRL", "HVLTREC", "hvlt_discrimination", "hvlt_retention", "hvlt_immediaterecall",
    "TMT_A", "TMT_B", "SDMTOTAL", "clockdraw",
    "scopa", "scopa_gi", "scopa_pm", "scopa_sex", "scopa_ur", "orthostasis",
    "quip_any", "quip_eat", "quip_sex", "quip_walk", "quip_hobby"]

# Bottom-Up Analysis (Hierarchical) Defining 
motor_signs = {
    # Bradykinesia
    "bradykinesia": [f for f in df.columns if any(k in f for k in ["SP_", "CAD_", "STR_T_", "TUG1_DUR", 
                                                                   "TUG2_DUR"])],
    # Hypokinesia – amplitude
    "hypokinesia": [f for f in df.columns if any(k in f for k in ["RA_AMP", "LA_AMP", "T_AMP"])],
    # Asymmetry – lateral imbalance
    "asymmetry": [f for f in df.columns if any(k in f for k in ["ASA_", "ASYM_IND_", "SYM_", "STEP_SYM_"])],
    # Dyscoordination
    "dyscoordination": [f for f in df.columns if any(k in f for k in ["TRA_", "STEP_SYM_", "TUG1_TURNS_DUR", 
                                                                      "TUG2_TURNS_DUR"])],
    # Smoothness Loss
    "smoothness_loss": [f for f in df.columns if "JERK_" in f],
    # Variability
    "variability": [f for f in df.columns if any(k in f for k in ["STR_CV_", "STEP_REG_"])],
    # Postural Instability
    "postural_instability": [f for f in df.columns if any(k in f for k in ["SW_VEL_", "SW_PATH_", "SW_FREQ_", 
                                                                       "SW_JERK_"])]
}

level3_df = pd.DataFrame(index=df.index) # Creating a new df for level 3

# Aggregating features into composites one composite numeric score per participant per motor sign
print(level3_df.isna().sum())
for sign, feats in motor_signs.items():
    valid_feats = [f for f in feats if f in df.columns]
    if len(valid_feats) > 0:
        level3_df[sign] = df[valid_feats].mean(axis=1) # Averages all those related features row-by-row
# Merge with main dataframe for modeling
df = pd.concat([df, level3_df], axis=1)

feature_groups = {"gait": gait_features, "demographic": demographic_features,"clinical": clinical_features, "combined": gait_features + demographic_features + clinical_features,
                 "motor_signs": list(level3_df.columns), "hierarchical_combined": gait_features + demographic_features + clinical_features + list(level3_df.columns)}
```

```python
# Training and evaluation function 
def run_model(X, y, name="Model"):
    # Imputing missing values using mean, then training a histogram gradient boosting classifier
    pipeline = Pipeline([("imputer", SimpleImputer(strategy="mean")),("clf", HistGradientBoostingClassifier(random_state=42))])
    
    #determining the smallest class size to avoid creating too many folds
    min_class_size = y.value_counts().min()
    n_splits = min(5, min_class_size)
    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    # Evaluate model with AUC using cross-validation
    scores = cross_val_score(pipeline, X, y, cv=cv, scoring="roc_auc")
    print(f"{name:12}: AUC = {scores.mean():.3f} ± {scores.std():.3f}")
    
    # Summary statistics 
    mean_auc = scores.mean()
    std_auc = scores.std()
    pipeline.fit(X, y)
    return pipeline, mean_auc, std_auc

# Training + evaluating over each feature group 
for group_name, features in feature_groups.items():
    print(group_name)
    X = df[features]
    model, mean_auc, std_auc = run_model(X, y, name=f"{group_name} Model")

    # Generating SHAP explanations for the trained calssifier
    explainer = shap.Explainer(model.named_steps["clf"], X)
    shap_values = explainer(X)
    
    # SHAP Summary Plot of each feature across the dataset
    shap.summary_plot(shap_values, X, max_display=20, show=True)
    # SHAP Bar Plot
    shap.plots.bar(shap_values, max_display=20, show=True)
    
# Seperately running the model for the special combined feature sets
for group_name in ["motor_signs", "hierarchical_combined"]:
    X = df[feature_groups[group_name]]
    model, mean_auc, std_auc = run_model(X, y, name=f"{group_name} Model")
```

- Gait alone is not a sufficient classifier, but it's an important objective, remote-accessible modality.
- Integration boosts robustness and may allow partial prediction in remote settings when clinical evals aren't available.
- Motor signs, which are derived from gait features have better predictive performance than the raw gait features

- Before interpreting results, we must rule out leakage or circularity, especially with clinical scores that could be correlated with diagnosis or collected post-CSFSAA.

```python
# Clinical Subgroups
cognitive_features = [
    "moca", "MCI_testscores", "cogstate",
    "HVLTFPRL", "HVLTREC", "hvlt_discrimination", "hvlt_retention", "hvlt_immediaterecall",
    "TMT_A", "TMT_B", "SDMTOTAL", "clockdraw"]

mood_behavior_features = [
    "gds", "ess", "stai", "stai_state", "stai_trait"]

updrs_part2_features = [
    "NP1ANXS", "NP1APAT", "NP1COG", "NP1DPRS", "NP1FATG", "NP1HALL", "NP1DDS"]

autonomic_features = [
    "scopa", "scopa_gi", "scopa_pm", "scopa_sex", "scopa_ur", "orthostasis"]

impulse_control_features = [
    "quip_any", "quip_eat", "quip_sex", "quip_walk", "quip_hobby"]

def evaluate_group(features, label, name):
    X = df[features]
    run_model(X, label, name)

# Run each group
evaluate_group(cognitive_features, df["label"], "Cognitive")
evaluate_group(mood_behavior_features, df["label"], "Mood/Behavior")
evaluate_group(updrs_part2_features, df["label"], "UPDRS Part II")
evaluate_group(autonomic_features, df["label"], "Autonomic")
evaluate_group(impulse_control_features, df["label"], "Impulse Control")
```

```python
# Converting SHAP values into a dataframe (each row = one sample, each column = SHAP value for a feature)
shap_df = pd.DataFrame(shap_values.values, columns=X.columns)

# Standardize SHAP values before UMAP 
shap_scaled = StandardScaler().fit_transform(shap_df)

# Initializing and running UMAP dimensionality reduction
reducer = umap.UMAP(n_neighbors=10, min_dist=0.3, metric='euclidean', random_state=42)
embedding = reducer.fit_transform(shap_scaled)

# Plotting UMAP embedding
plt.figure(figsize=(8, 6))
sns.scatterplot(x=embedding[:, 0], y=embedding[:, 1], 
    hue=y.map({0: "Negative", 1: "Positive"}),
    palette={"Negative": "gray", "Positive": "orange"},
    s=70, alpha=0.9)
plt.title("UMAP of SHAP Values (CSFSAA Label)")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend(title="CSFSAA")
plt.tight_layout()
plt.show()
```

```python
# Scaling SHAP values before clustering
scaler = StandardScaler()
shap_scaled = scaler.fit_transform(shap_values.values)


# UMAP plot colored by SHAP of a key feature
feature = "R_JERK_U" # Change to visualize other features
if feature in X.columns:
    plt.figure(figsize=(8, 6))
    # Plotting the existing UMAP embedding, coloring points by SHAP value
    plt.scatter(embedding[:, 0], embedding[:, 1], c=shap_df[feature], cmap="coolwarm", s=70)
    plt.colorbar(label=f"SHAP Value: {feature}")
    plt.title(f"UMAP Colored by SHAP of '{feature}'")
    plt.xlabel("UMAP 1")
    plt.ylabel("UMAP 2")
    plt.tight_layout()
    plt.show()
else:
    print(f"Feature '{feature}' not found in SHAP values.")

# Cluster SHAP profiles using K-means and UMAP
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(shap_scaled)

# Cluster the SHAP value patterns into 3 groups
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(shap_scaled)

# Plot UMAP clusters
plt.figure(figsize=(8, 6))

# Plot each cluster as a different color
for cluster_id in np.unique(clusters):
    idx = clusters == cluster_id
    plt.scatter(
        embedding[idx, 0], embedding[idx, 1],
        label=f"Cluster {cluster_id}", s=70)

# Overlay CSFSAA markers (ground truth) on top of clusters
for csfsaa_label in [0, 1]:
    idx = y == csfsaa_label
    plt.scatter(
        embedding[idx, 0], embedding[idx, 1],
        marker='x', s=100,
        color='black' if csfsaa_label == 1 else 'gray',
        label=f"CSFSAA = {csfsaa_label}")

plt.title("UMAP: SHAP-Based Clusters + CSFSAA Overlay")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.legend()
plt.show()
```

## Spectral Co-Clustering

```python
# Selecting and scaling the combined feature group
X = df[feature_groups["hierarchical_combined"]]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Select only top SHAP features
X_top = df[top_features]

# Imputing missing values using mean and standardize
X_imputed = SimpleImputer(strategy="mean").fit_transform(X_top)
X_scaled = StandardScaler().fit_transform(X_imputed)

# Apply spectral co-clustering (groups of patients + groups of features)
model = SpectralCoclustering(n_clusters=3, random_state=0)
model.fit(X_scaled)

# Reorder rows and columns for grouped heatmap according to cluster assignments
row_order = np.argsort(model.row_labels_)
col_order = np.argsort(model.column_labels_)

# Re-ordering matrix for visualization
X_ordered = X_scaled[row_order][:, col_order]

# CSFSAA annotation sorting
csfsaa_sorted = df.loc[df.index[row_order], "label"].astype(str).reset_index(drop=True)

# Extracting cluster assignments for rows (patients) and columns (features)
row_clusters = model.row_labels_[row_order]
col_clusters = model.column_labels_[col_order]

# Calculating CSFSAA-positive % per cluster
df_clustered = df.iloc[row_order].reset_index(drop=True)
df_clustered["Cluster"] = row_clusters
csfsaa_percent = df_clustered.groupby("Cluster")["label"].mean() * 100

# Plotting Heatmap 
plt.figure(figsize=(14, 10))
ax = sns.heatmap(X_ordered, cmap="coolwarm", cbar=False) 

# Drawing cluster boundaries + annotations
unique_clusters = np.unique(row_clusters)
for cluster_id in unique_clusters:
    idx = np.where(row_clusters == cluster_id)[0]
    start, end = idx[0], idx[-1]
    
    # Horizontal boundary line for each patient cluster
    plt.hlines([end + 1], xmin=0, xmax=X_ordered.shape[1], color="black", linewidth=1.5)

    # Subgroup Label
    plt.text(-3, (start + end) / 2, f"Patient\nSubgroup {cluster_id + 1}:", va='center', ha='right', fontsize=9, fontweight='bold')
    
    # CSFSAA percent annotation
    pct = csfsaa_percent[cluster_id]
    plt.text(-1.5, (start + end) / 2, f"CSFSAA:\n{pct:.1f}%", va='center', ha='right', fontsize=9)

# Draw feature cluster boundaries (vertical bondaries)
unique_col_clusters = np.unique(col_clusters)
for cid in unique_col_clusters:
    idx = np.where(col_clusters == cid)[0]
    start, end = idx[0], idx[-1]
    plt.vlines([end + 1], ymin=0, ymax=X_ordered.shape[0], color="black", linewidth=1.5)

    # Column label in middle of block
    plt.text((start + end) / 2, -4, f"{cid + 1}", ha='center', va='bottom', fontsize=10)

# Axis Labels and Ticks
plt.xticks(ticks=np.arange(len(col_order)), 
           labels=np.array(top_features)[col_order], 
           rotation=90, fontsize=8)
plt.yticks([]) # Hide ticks since we annotate instead

plt.title("Spectral Co-Clustering Heatmap (Top SHAP Features)")
plt.xlabel("")
plt.ylabel("")

cbar = plt.colorbar(ax.collections[0])
cbar.set_label("Z", rotation=0, labelpad=10)
plt.tight_layout()
plt.show()
```

## Balanced Sampling + Evaluation

```python
# Mean imputation for missing values
imputer = SimpleImputer(strategy="mean")

# SMOTE for class balancing within each training fold
smote = SMOTE(random_state=42, sampling_strategy='auto')

# 5-fold stratified CV to preserve label distribution
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = []

# Looping through each feature group 
for group_name, feature_list in feature_groups.items():
    print(f"\nRunning {group_name.upper()} model...")

    # Extract features and target label
    X = df[feature_list]
    y = df["label"]

    auc_scores, pr_auc_scores = [], []
    
    # Cross validation: split, impute, resample, train
    for train_idx, test_idx in cv.split(X, y):
        # Splitting data
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        # Impute missing values
        X_train_imp = imputer.fit_transform(X_train)
        X_test_imp = imputer.transform(X_test)

        # Apply SMOTE
        X_res, y_res = smote.fit_resample(X_train_imp, y_train)

        # Create model pipeline
        model = Pipeline([
            ("scaler", StandardScaler()),
            ("clf", HistGradientBoostingClassifier(random_state=42))])
        model.fit(X_res, y_res)

        # Predicting probabilities
        y_pred = model.predict_proba(X_test_imp)[:, 1]

        # Computing metrics
        auc_scores.append(roc_auc_score(y_test, y_pred))
        precision, recall, _ = precision_recall_curve(y_test, y_pred)
        pr_auc_scores.append(auc(recall, precision))

    # Storing fold-averaged metrics for the current feature group 
    results.append({
        "Group": group_name,
        "ROC-AUC Mean": np.mean(auc_scores),
        "ROC-AUC SD": np.std(auc_scores),
        "PR-AUC Mean": np.mean(pr_auc_scores),
        "PR-AUC SD": np.std(pr_auc_scores)
    })

# Create results DataFrame
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by="ROC-AUC Mean", ascending=False).reset_index(drop=True)

print("\n\nBalanced SMOTE Results Across Feature Groups")
print(results_df.round(3))
```

```python
from scipy.spatial.distance import cdist
distances = cdist(X_train_imp, X_res[:len(X_train_imp)])
print("Mean distance between original and synthetic:", distances.mean())
```

To assess class imbalance effects, SMOTE-based oversampling was applied within cross-validation folds.
While models trained with SMOTE maintained strong performance overall, the feature group ranking inverted (clinical > combined > hierarchical_combined).
This inversion coincided with a large average distance (≈ 130.8) between real and synthetic samples in the hierarchical embedding space, suggesting that SMOTE interpolation generates non-biological synthetic points far from the data manifold.

In contrast, lower-dimensional clinical features allow smoother interpolation and thus appear to benefit more from SMOTE.
This observation reinforces the interpretation that hierarchical embeddings capture nonlinear, structured variability that cannot be approximated through simple linear resampling.

Consequently, our primary results rely on the non-SMOTE models, which preserve the natural data distribution. SMOTE results are reported as a sensitivity analysis, demonstrating that predictive signal remains strong under balanced training conditions.
